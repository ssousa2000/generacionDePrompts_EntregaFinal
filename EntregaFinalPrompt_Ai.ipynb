```python
Trabajo Final:
Roomies - Gestión de Tareas para Compañeros de Cuarto con IA
Resumen:
Roomies es una aplicación que distribuye equitativamente las tareas del hogar entre compañeros de cuarto. Utilizando OpenAI GPT-4, la app genera una lista de tareas personalizada basada en la cantidad de personas y la frecuencia de limpieza requerida. Además, mediante DALL-E, se crea una representación visual de la distribución de tareas para mejorar la experiencia del usuario. Si un compañero no cumple con su tarea, deberá pagar una compensación a otro usuario a través de plataformas de pago como Yape o Plin.
Índice
Introducción
Objetivos
Metodología
Herramientas y Tecnologías
Implementación
Resultados
Conclusiones
Referencias
1. Introducción
En la vida compartida con compañeros de cuarto, la distribución de tareas domésticas suele generar conflictos. La falta de equidad en la asignación de tareas o el incumplimiento de las mismas puede causar fricciones en la convivencia. Roomies busca resolver este problema automatizando la asignación de tareas y aplicando consecuencias económicas en caso de incumplimiento.
2. Objetivos
Crear un sistema que asigne automáticamente tareas del hogar de forma equitativa.
Utilizar GPT-4 para generar listas de tareas optimizadas.
Implementar DALL-E para generar una visualización de la distribución de tareas.
Automatizar la penalización de tareas incumplidas con un sistema de pagos.
3. Metodología
Recopilación de Datos: Se ingresa la cantidad de roommates y las tareas necesarias en la casa.
Generación de Tareas: GPT-4 distribuye las tareas de forma equitativa y sugiere nuevos quehaceres si es necesario.
Visualización con DALL-E: Se genera una imagen que muestra la asignación de tareas de cada roommate.
Registro de Cumplimiento: Se utiliza un sistema de verificación donde cada usuario marca su tarea como completada.
Penalización Automática: Si un usuario no completa su tarea en el tiempo asignado, se le descuenta dinero a través de Yape o Plin y se transfiere a otro usuario que haya cumplido sus tareas.
4. Herramientas y Tecnologías
Python para la lógica del sistema.
OpenAI GPT-4 API para generar listas de tareas personalizadas.
OpenAI DALL-E API para crear representaciones visuales.
Sistema de pagos (Yape o Plin) para gestionar penalizaciones.
5. Implementación
5.1 Configuración de la API de OpenAI
import openai
openai.api_key = "TU_CLAVE_API"
5.2 Generación de Tareas con GPT-4
def generar_tareas(roommates):
    prompt = f"""
    Genera una lista de tareas domésticas para {roommates} compañeros de cuarto.
    Distribúyelas equitativamente y especifica frecuencia de ejecución.
    """
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "Eres un organizador experto en tareas del hogar."},
                  {"role": "user", "content": prompt}]
    )
    return response["choices"][0]["message"]["content"]
5.3 Generación de Imagen con DALL-E
def generar_imagen(tareas):
    response = openai.Image.create(
        prompt=f"Ilustración de la asignación de tareas en un hogar con {tareas}",
        n=1,
        size="512x512"
    )
    return response['data'][0]['url']
5.4 Implementación de Penalizaciones
def verificar_tareas(roommate, tareas_completadas):
    if roommate not in tareas_completadas:
        print(f"{roommate} no completó su tarea. Se descontará dinero por Yape o Plin.")
        # Integración con Yape o Plin (simulada)
        return True
    return False
6. Resultados
La implementación de Roomies genera una asignación automática de tareas domésticas, mejorando la convivencia entre compañeros de cuarto. Además, la visualización con DALL-E ayuda a que cada usuario vea claramente su responsabilidad. Finalmente, el sistema de penalización motiva a los usuarios a cumplir con sus tareas, evitando conflictos.
7. Conclusiones
La IA facilita la organización del hogar al distribuir tareas de manera equitativa.
La integración de DALL-E ayuda a la visualización de las asignaciones.
La implementación de penalizaciones económicas fomenta la responsabilidad en la convivencia.
8.(1. Técnicas de Fast Prompting utilizadas
En la generación de prompts para GPT-4 y DALL-E, se utilizaron las siguientes técnicas de fast prompting:

- Estructuración clara de los prompts**: Para la generación de tareas, se utilizó un prompt que indica explícitamente que las tareas deben ser distribuidas equitativamente y con una frecuencia definida.
- Instrucciones concisas y detalladas**: Para DALL-E, el prompt describe la escena de manera clara, indicando elementos clave como "compañeros de cuarto interactuando en un ambiente hogareño con una lista de tareas visible en una aplicación móvil".
- Iteración y refinamiento**: Se realizaron pruebas con diferentes formulaciones del prompt para asegurar que la IA genere listas de tareas completas y una imagen que represente adecuadamente la distribución de responsabilidades en la convivencia.

2. Viabilidad del proyecto
Este proyecto es viable en términos técnicos y de recursos disponibles. OpenAI GPT-4 permite la generación de listas de tareas automatizadas de manera eficiente, mientras que DALL-E ayuda a la visualización de las asignaciones.

En cuanto a costos, el uso de la API de OpenAI tiene costos variables dependiendo del número de consultas realizadas. Sin embargo, este gasto puede optimizarse ajustando los prompts para obtener respuestas más precisas y reduciendo la cantidad de llamadas a la API. Además, el uso de plataformas de pago como Yape o Plin para la penalización de tareas incumplidas es factible, ya que estas herramientas son ampliamente utilizadas y permiten transacciones rápidas entre usuarios.)

9. Referencias
Documentación de OpenAI GPT-4 API
Documentación de OpenAI DALL-E API
Guías de integración de pagos con Yape y Plin


https://drive.google.com/file/d/1RwcEALVbsNkWfik7P9jzd1g2HkgmJbAX/view?usp=sharing
from IPython.display import display, Image
display(Image(url=https://drive.google.com/file/d/1RwcEALVbsNkWfik7P9jzd1g2HkgmJbAX/view?usp=sharing))

```
